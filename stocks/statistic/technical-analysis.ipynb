```python
import matplotlib.pyplot as plt
import mpl_finance as mpf
#from matplotlib.pylab import data2num
#import datetime

plt.rcParams['font.sans-serif']=['SimHei']
#df.reset_index(inplace=True)
df.set_index('trade_date', inplace=True)
df.index = df.index.format(formatter=lambda x:x.strftime('%Y-%m-%d'))
df
```

```python
import numpy as np
import talib

sma_10 = talib.SMA(np.array(df['close']), 10)
sma_12 = talib.SMA(np.array(df['close']), 12)
sma_30 = talib.SMA(np.array(df['close']), 30)
fig = plt.figure(figsize=(24, 20))
# add_axes([left, bottom, width, height]): left,bottom是相对于左下角原点的相对位置，width,height是相对于figure的宽高比例
ax = fig.add_axes([0,0.3,1,0.4])
ax2 = fig.add_axes([0,0.2,1,0.1])
ax3 = fig.add_axes([0,0,1,0.2])
ax.set_xticks(range(0, len(df.index), 80))
ax.set_xticklabels(df.index[::80])
mpf.candlestick2_ochl(ax, df['open'], df['close'], df['high'],
                      df['low'], width=0.6, colorup='r', colordown='g', alpha=0.75)
#plt.rcParams['font.sans-serif']=['Microsoft JhengHei'] 
ax.plot(sma_10, label='10日均線')
ax.plot(sma_12, label='12日均線')
ax.plot(df['MA12'], label='12日均线N')
ax.plot(sma_30, label='30日均線')

ax2.plot(df['BIAS'], label='乖離率')
ax2.plot(df['PSY'], label='心理線')
ax2.plot(df['RSI'], label='相對強弱指標')
#ax2.plot(df['AR'], label='買賣氣勢指標')
ax2.plot(df['BR'], label='買賣意願指標')
ax2.plot(df['WMS'], label='威廉指標')
ax2.set_xticks(range(0, len(df.index), 80))
ax2.set_xticklabels(df.index[::80])

mpf.volume_overlay(ax3, df['open'], df['close'], df['vol'], colorup='r', colordown='g', width=0.5, alpha=0.8)
ax3.plot(df['MV'], label='平均成交量')
ax3.plot(df['OBV'], label='能量潮')
ax3.set_xticks(range(0, len(df.index), 80))
ax3.set_xticklabels(df.index[::80])


ax.legend();
ax2.legend();
ax3.legend();
```
